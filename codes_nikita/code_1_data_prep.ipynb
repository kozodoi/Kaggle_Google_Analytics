# 2. FUNCTIONS

```python
##### FUNCTION FOR LOADING DATA WITH JSON VALUES
def read_csv_with_json(path, json_cols, nrows = None):
        
    # import data frame
    df = pd.read_csv(path, 
                     converters = {column: json.loads for column in json_cols}, 
                     dtype = {'fullVisitorId': 'str'}, # Important!!
                     nrows = nrows)
    
    # extract values
    for column in json_cols:
        column_as_df = json_normalize(df[column])
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns]
        df = df.drop(column, axis = 1).merge(column_as_df, right_index = True, left_index = True)

    # return data
    print(f"Loaded {os.path.basename(path)}. Shape: {df.shape}")
    return df
```

# 3. IMPORT

```python
# JSON columns
json_cols = ['device', 'geoNetwork', 'totals', 'trafficSource']

# import data
train = read_csv_with_json("../data/raw/train.csv", json_cols = json_cols)
test  = read_csv_with_json("../data/raw/test.csv",  json_cols = json_cols)
```

# 4. MERGER

```python
### MERGE DATA

# correct targets
train["target"] = train["totals.transactionRevenue"].fillna(0)
test["totals.transactionRevenue"] = None
del train["totals.transactionRevenue"]

# align columns
del train["trafficSource.campaignCode"]
train = train.reindex_axis(sorted(train.columns), axis = 1)
test  = test.reindex_axis(sorted(test.columns),   axis = 1)

# concatenate
df = pd.concat([train, test], axis = 0)

# check shape
print(df.shape)
```

```python
# remove all features with single value
print(df.shape)
df = df.loc[:,df.nunique(dropna = False) != 1]
print(df.shape)
```
